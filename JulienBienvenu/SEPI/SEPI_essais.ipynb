# Mapping with MAPK

## Compiling MAPK

```biocham
option(negation_first:no).
option(quadratic_reduction:sat_species).

clear_model.
parameter(k=3.0, r=2.0).
compile_function(h = r*x^5/(k^5+x^5), x).
parameter(input = 2.0).
export_biocham("hill5.bc").
```

## Uploading MAPK

```biocham
download_curated_biomodel(9).
```

## Searching for SEPI

The hill function of order 5 have been saved in the `hill5.bc` and we upload the model of Huang and Ferrel (model 009 on biomodel). To check if a SEPI exist woth the MAPK signalling cascade,  and launch the following command.

We enforce the mapping in such a way that the biological input `E1` is mapped on our input `x` and similarly for the output. In the model of Hang & Ferrel, both `PP_K` or `KPase_PP_K` may be seen as output. Both choice leads to the existence of a SEPI.

```biocham
search_reduction("BIOMD0000000009.xml", "hill5.bc",
    mapping_restriction:[E1->x, PP_K->h]).
```
