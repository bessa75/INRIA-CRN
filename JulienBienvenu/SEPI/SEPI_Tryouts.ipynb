GluONe -> Resorufin = glucose$\wedge$acetone et NADH = glucose $\wedge$(1-acetone) <br>
LacOH  -> NADH = EtOH et ABTSOX = Lactate$\wedge$(1-EtOH) <br>
GluNOx -> DAFF = glucose$\wedge$NOx et NADH = glucose$\wedge$(1-NOx)

```biocham
clear_model.
add_function( resorufin = (acetone * glucose), NADH = (glucose * (1 - acetone))).
list_model.
draw_reactions.
```

```biocham
clear_model.
add_function( NADH = EtOH, ABTSOX = (1-EtOH)*Lactate).
list_model.
draw_reactions.
```

```biocham
clear_model.
add_function( NADH = glucose*(1-NOx), DAFF = glucose*NOx).
list_model.
draw_reactions.
```
